```python
import pandas as pd
```

```python
df=pd.read_excel('HHRA.xlsx')
df.head(84)
```

```python
def exmp(c):
    adi_NO3_infant = (c * 0.31 * 1 * 365) / (6.9 * 365)
    inhi_NO3_infant = adi_NO3_infant / 1.6
    return inhi_NO3_infant
x=exmp (df.NO3)
df['INHI NO3 infant']=x
```

```python
df
```

```python
def exmp(c):
    adi_NO3_children = (c * 0.68 * 12 * 365) / (18.7 * 4380)
    inhi_NO3_children = adi_NO3_children / 1.6
    return inhi_NO3_children
x=exmp (df.NO3)
df['INHI NO3 children']=x
```

```python
df
```

```python
def exmp(c):
    adi_NO3_adult = (c * 2.5 * 64 * 365) / (57.5 * 23360)
    inhi_NO3_adult = adi_NO3_adult / 1.6
    return inhi_NO3_adult
x=exmp (df.NO3)
df['INHI NO3 adult']=x
```

```python
df
```

```python
def exmp(c):
    adi_F_infant = (c * 0.31 * 1 * 365) / (6.9 * 365)
    inhi_F_infant = adi_F_infant / 0.06
    return inhi_F_infant
x=exmp (df.F)
df['INHI F infant']=x
```

```python
df
```

```python
def exmp(c):
    adi_F_children = (c * 0.68*12*365) / (18.7*4380)
    inhi_F_children = adi_F_children / 0.06
    return inhi_F_children
x=exmp (df.F)
df['INHI F children']=x
```

```python
df
```

```python
def exmp(c):
    adi_F_adult = (c * 2.5*64*365) / (57.5*23360)
    inhi_F_adult = adi_F_adult / 0.06
    return inhi_F_adult
x=exmp (df.F)
df['INHI F adult']=x
```
