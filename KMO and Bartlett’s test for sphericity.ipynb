```python
!pip install factor-analyzer
import sys
print(sys.executable)


import pandas as pd
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_kmo, calculate_bartlett_sphericity


file_path = "waterquality.xlsx"  
df = pd.read_excel(file_path)


parameters = ['pH','EC','TDS','TH','Na','K','Ca','Mg','SO4','NO3','Cl','F','HCO3']
data = df[parameters]


if data.isnull().sum().sum() > 0:
    print("Warning: Missing values found. Fill or drop them before analysis.")
else:
   
    kmo_all, kmo_model = calculate_kmo(data)
    print(f"Overall KMO: {kmo_model:.3f}")
    print("KMO per variable:")
    for var, kmo_val in zip(parameters, kmo_all):
        print(f"{var}: {kmo_val:.3f}")

    
    chi_square_value, p_value = calculate_bartlett_sphericity(data)
    print(f"\nBartlett's Test Chi-square: {chi_square_value:.3f}")
    print(f"Bartlett's Test p-value: {p_value:.3e}")

```
