```python
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import os

# === Step 1: Load your Excel file ===
df = pd.read_excel('wqi _annova.xlsx')  # Change if your file name is different

# === Step 2: Run ANOVA ===
groups = df.groupby('cluster')['wqi'].apply(list)
f_stat, p_value = stats.f_oneway(*groups)

print(f"\nANOVA F-statistic: {f_stat:.3f}")
print(f"ANOVA p-value: {p_value:.3f}")

if p_value < 0.05:
    print("‚úÖ There is a significant difference in WQI between clusters.")
else:
    print("‚ùå No significant difference in WQI between clusters.")

# === Step 3: Create and Save Boxplot ===
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

sns.boxplot(x='cluster', y='wqi', data=df, palette='Set2')

plt.title('WQI Distribution by Cluster', fontsize=14)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Water Quality Index (WQI)', fontsize=12)
plt.tight_layout()

# --- Save as high-resolution TIFF on Desktop ---
desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")
save_path = os.path.join(desktop_path, "WQI_boxplot.tiff")
plt.savefig(save_path, format='tiff', dpi=300)

print(f"\nüìÅ Plot saved successfully to: {save_path}")

# Show plot
plt.show()
```
