```python
import pandas as pd
from scipy.stats import levene

df = pd.read_excel("waterquality.xlsx")


group_col = "cluster"  
indices = ["wqi", "pig"]

print("Levene’s Test for Homogeneity of Variance\n")

for index in indices:
    
    groups = [
        df[df[group_col] == g][index].dropna()
        for g in df[group_col].unique()
    ]
    
    stat, p = levene(*groups)
    
    print(f"{index}:")
    print(f"  Levene statistic = {stat:.4f}")
    print(f"  p-value          = {p:.4f}\n")
```
