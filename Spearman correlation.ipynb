```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel("waterquality.xlsx")

params = [
    "pH", "EC", "TDS", "TH", "Na", "K", "Ca", "Mg",
    "SO4", "NO3", "Cl", "F", "HCO3"
]

corr = df[params].corr(method="spearman")

plt.figure(figsize=(10, 8))
plt.imshow(corr, interpolation="nearest")
plt.colorbar(label="Spearman correlation ()")

plt.xticks(range(len(params)), params, rotation=45, ha="right")
plt.yticks(range(len(params)), params)

for i in range(len(params)):
    for j in range(len(params)):
        plt.text(j, i, f"{corr.iloc[i, j]:.2f}",
                 ha="center", va="center", fontsize=8)

plt.title("Spearman Rank Correlation Matrix")

plt.tight_layout()
plt.savefig(
    "Spearman_Correlation_Matrix.tiff",
    format="tiff",
    dpi=300
)

plt.show()
```
