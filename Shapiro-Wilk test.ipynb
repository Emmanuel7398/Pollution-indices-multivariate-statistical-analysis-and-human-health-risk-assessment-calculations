```python
import pandas as pd
from scipy.stats import shapiro


df = pd.read_excel("waterquality.xlsx")


parameters = [
    "pH", "EC", "TDS", "TH", "Na", "K", "Ca", "Mg",
    "SO4", "NO3", "Cl", "F", "HCO3"
]

print("Shapiro–Wilk Normality Test Results\n")
print("{:<10} {:<10} {:<15}".format("Parameter", "W-value", "p-value"))
print("-" * 35)


for param in parameters:
    data = df[param].dropna()  
    stat, p = shapiro(data)
    print("{:<10} {:<10.4f} {:<15.4f}".format(param, stat, p))
```
