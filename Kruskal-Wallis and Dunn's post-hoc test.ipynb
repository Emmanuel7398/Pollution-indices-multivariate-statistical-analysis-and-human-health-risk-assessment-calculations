```python

!pip install scikit-posthocs

import pandas as pd
import scikit_posthocs as sp
from scipy.stats import kruskal


df = pd.read_excel("waterquality.xlsx")


df['cluster'] = df['cluster'].astype(str)


groups_wqi = [group["wqi"].values for name, group in df.groupby("cluster")]
H_wqi, p_wqi = kruskal(*groups_wqi)

print("Kruskal–Wallis Test for WQI")
print(f"H statistic = {H_wqi:.4f}")
print(f"p-value     = {p_wqi:.4f}")

if p_wqi < 0.05:
    print("\nDunn’s Post-hoc Test for WQI (Bonferroni corrected)")
    dunn_wqi = sp.posthoc_dunn(df,
                               val_col='wqi',
                               group_col='cluster',
                               p_adjust='bonferroni')
    print(dunn_wqi)
else:
    print("\nNo significant difference in WQI. Dunn’s test not required.")


groups_pig = [group["pig"].values for name, group in df.groupby("cluster")]
H_pig, p_pig = kruskal(*groups_pig)

print("\n\nKruskal–Wallis Test for PIG")
print(f"H statistic = {H_pig:.4f}")
print(f"p-value     = {p_pig:.4f}")

if p_pig < 0.05:
    print("\nDunn’s Post-hoc Test for PIG (Bonferroni corrected)")
    dunn_pig = sp.posthoc_dunn(df,
                               val_col='pig',
                               group_col='cluster',
                               p_adjust='bonferroni')
    print(dunn_pig)
else:
    print("\nNo significant difference in PIG. Dunn’s test not required.")

```
