```python
import pandas as pd

# === Step 1: Load the Excel file ===
file_path = 'CAI.xlsx'  # üîÅ Replace with your actual file path
df = pd.read_excel(file_path)

# === Step 2: Check required columns ===
required_cols = ['Na', 'K', 'SO4', 'NO3', 'Cl', 'HCO3']
missing = [col for col in required_cols if col not in df.columns]
if missing:
    raise ValueError(f"Missing columns: {missing}")

# === Step 3: Calculate CAI1 and CAI2 ===
df['CAI1'] = (((df['Na'] + df['K'])-df['Cl']) / df['Cl'])
df['CAI2'] = (df['Na'] + df['K'] - df['Cl']) / (df['SO4'] + df['HCO3'] + df['NO3'])

# === Step 4: Save to new Excel file ===
output_file = 'CAI_results.xlsx'
df.to_excel(output_file, index=False)

print(f"‚úÖ CAI1 and CAI2 calculated and saved to '{output_file}'")
```
